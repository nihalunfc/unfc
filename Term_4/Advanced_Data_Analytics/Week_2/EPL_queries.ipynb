```python
%sql
-- getting top 10 scorers
SELECT 
PlayerName, 
Goals 
FROM damo630.english_league.players_premier  -- you may need to change your path here
ORDER BY Goals DESC LIMIT 10;
```

```python
%sql
-- getting teams names and points

SELECT 
  t.TeamName, 
  s.Points 
FROM damo630.english_league.standings_premier s
JOIN damo630.english_league.teams_premier t 
  ON s.TeamID = t.TeamID 
ORDER BY s.Points DESC;
```
