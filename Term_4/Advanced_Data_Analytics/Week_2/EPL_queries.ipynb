```python
# 1. Read directly from your new Databricks Catalog tables
df_players = spark.table("default.players_premier")
df_standings = spark.table("default.standings_premier")
df_teams = spark.table("default.teams_premier")

# 2. Register them as Temp Views so your %sql cells work
df_players.createOrReplaceTempView("players")
df_standings.createOrReplaceTempView("standings")
df_teams.createOrReplaceTempView("teams")

print("Tables successfully loaded and views are ready for querying!")
```

```python
%sql
SELECT PlayerName, Goals 
FROM players 
ORDER BY Goals DESC 
LIMIT 10;
```

```python
%sql
SELECT t.TeamName, s.Points 
FROM standings s
JOIN teams t ON s.TeamID = t.TeamID 
ORDER BY s.Points DESC;
```
