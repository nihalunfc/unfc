```python
import time
import random

def stream_generator():
    while True:
        yield {"timestamp": time.time(), "value": random.randint(0, 100)}

stream = stream_generator()
for _ in range(5):
    print(next(stream))
    time.sleep(1)
```

```python
from collections import deque
import random
import time

window = deque(maxlen=10)  # Holds last 10 items

for _ in range(20):
    value = random.randint(0, 100)
    window.append(value)
    print("Current window:", list(window))
    time.sleep(0.5)  # Simulates time delay in stream
```

```python
def tumbling_window(stream, window_size=5):
    buffer = []
    for item in stream:
        buffer.append(item)
        if len(buffer) == window_size:
            print("Window batch:", buffer)
            buffer.clear()

# Simulated stream
stream = (random.randint(0, 50) for _ in range(20))
tumbling_window(stream, window_size=5)
```

```python
import math
import time

def exponential_decay(values, timestamps, lambda_=0.1):
    now = time.time()
    weighted_sum = 0
    for v, t in zip(values, timestamps):
        age = now - t
        weight = math.exp(-lambda_ * age)
        weighted_sum += v * weight
    return weighted_sum

# Simulated time series
timestamps = [time.time() - i for i in range(5)]  # 5s ago to now
values = [10, 20, 30, 40, 50]

print("Exponentially weighted sum:", exponential_decay(values, timestamps))
```

```python
import random

def reservoir_sample(stream, k):
    reservoir = []
    for i, item in enumerate(stream):
        if i < k:
            reservoir.append(item)
        else:
            j = random.randint(0, i)
            if j < k:
                reservoir[j] = item
    return reservoir

# Simulated stream: numbers 1 to 100
stream = range(1, 101)
sample = reservoir_sample(stream, 5)
print("Reservoir sample:", sample)
```

```python
import random

def count_distinct_naive(stream):
    seen = set()
    for item in stream:
        seen.add(item)
    return len(seen)

# Example stream
stream = [random.randint(1, 1000) for _ in range(10000)]
print("Naive distinct count:", count_distinct_naive(stream))
```

```python
from datasketch import HyperLogLog
import random

# Initialize HLL with precision p (higher = more accurate, more memory)
hll = HyperLogLog(p=10)

# Simulated stream with many duplicates
stream = [random.randint(1, 10000) for _ in range(50000)]

for item in stream:
    hll.update(str(item).encode('utf-8'))  # Input must be bytes

print("HyperLogLog estimate:", int(hll.count()))
```

```python
# Naive count (exact)
true_distinct = len(set(stream))

# HyperLogLog estimate
hll_estimate = int(hll.count())

print("True distinct count:", true_distinct)
print("HyperLogLog estimate:", hll_estimate)
print("Relative error:", abs(hll_estimate - true_distinct) / true_distinct)
```

```python
from collections import deque
import random
import time

window = deque(maxlen=5)

for _ in range(15):
    value = random.randint(0, 100)
    window.append(value)
    avg = sum(window) / len(window)
    print(f"New value: {value}, Rolling average: {avg:.2f}")
    time.sleep(0.5)
```

```python
import math
import time
def exponential_decay(values, timestamps, lambda_=0.1):
    now = time.time()
    weighted_sum = 0
    total_weight = 0

    for v, t in zip(values, timestamps):
        age = now - t
        weight = math.exp(-lambda_ * age)
        weighted_sum += v * weight
        total_weight += weight

    return weighted_sum / total_weight if total_weight > 0 else 0
```

```python
now = time.time()
timestamps = [now - i for i in reversed(range(5))]  # ensures correct order: oldest to newest
values = [10, 20, 30, 150, 200]

print("λ=0.1:", exponential_decay(values, timestamps, lambda_=0.1))
print("λ=0.3:", exponential_decay(values, timestamps, lambda_=0.3))
print("λ=0.7:", exponential_decay(values, timestamps, lambda_=0.7))
```
