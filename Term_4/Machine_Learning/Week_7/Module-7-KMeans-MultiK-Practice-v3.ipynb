## In-Class Practice: Clustering & PCA (v3 with K-Means Visualization)

**New Option:** Now includes visualization of K-Means results with multiple values of k (2–6).

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Generate dataset
X, y_true = make_blobs(n_samples=500, centers=4, n_features=5, random_state=42)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA for 2D projection
pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X_scaled)

# Visualize K-Means with k = 2 to 6
fig, axes = plt.subplots(2, 3, figsize=(12, 8))
axes = axes.ravel()
for i, k in enumerate([2, 3, 4, 5, 6]):
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = km.fit_predict(X_scaled)
    axes[i].scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap='tab10', s=20)
    axes[i].set_title(f'K-Means (k={k})')
    axes[i].set_xticks([])
    axes[i].set_yticks([])
plt.tight_layout()
plt.show()
```
